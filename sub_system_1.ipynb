```python
tiger_img = cv2.imread('tiger.jpg',0) # reads image 'opencv-logo.png' as grayscale
#plt.imshow(tiger_img, cmap='gray')
img = cv2.resize(tiger_img,(500,500))
#image_cordinates = 
#print(tiger_img)

grid_image_array = []
def create_grid_boxes(img):
    px_mov = 50;
    x_max, y_max  = img.shape
    y_pos = 0
    x_pos = 0
    land_mark_counter = 0
    while y_pos < y_max:
        while x_pos < x_max:
            #1 Det här är sub-bilden
            grid_img = img[x_pos:(x_pos+px_mov), y_pos:(y_pos+px_mov)]
            for i in range(0,(len(X.columns)),2):
                x_cord = X.iloc[0][i]
                y_cord = X.iloc[0][i+1]
                if (x_cord !=0) and (y_cord !=0) and (x_pos <= x_cord <=(x_pos+px_mov)) and (y_pos <= y_cord <=(y_pos+px_mov)):
                    land_mark_counter += 1
                    grid_img_label = X.columns[i]
                    print('found landmark:')
                    print(X.columns[i])
                    print('at:')
                    print(x_cord,y_cord) 
                else : 
                    grid_img_label = "no_landmark"
                    
            #2 Här borde sub-bilden sparas med label i en array
            grid_image_array.append([grid_img,grid_img_label])
            x_pos += px_mov
            #plt.imshow(grid_img, cmap='gray')
            #print(grid_img)
            #return
        x_pos = 0
        y_pos += px_mov
    return(grid_image_array)


#Detta är arrayen med alla sub-bilder och labels
grid_array = create_grid_boxes(img)
```
