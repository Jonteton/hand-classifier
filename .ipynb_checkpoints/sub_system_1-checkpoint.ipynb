```python

grid_image_array = []

def create_grid_boxes(img, data):

    px_x_mov = 32
    px_y_mov = 24
    x_max = 640
    y_max = 480
    y_pos = 0
    x_pos = 0
    land_mark_counter = 0
    actuallandmarkcounter = 0
    while y_pos < y_max:
        while x_pos < x_max:
            grid_img_label = "no_landmark"
            #1 Det här är sub-bilden
            grid_img = img[y_pos:(y_pos+px_y_mov), x_pos:(x_pos+px_x_mov)]
            for i in range(0,(len(X.columns)),2):
                x_cord = data.iloc[0][i]
                y_cord = data.iloc[0][i+1]
                
                if (x_cord != 0) and (y_cord != 0) and (x_pos <= x_cord <= (x_pos+px_x_mov)) and (y_pos <= y_cord <= (y_pos+px_y_mov)):
                    land_mark_counter +=1
                    grid_img_label = data.columns[i]
                    #print('found landmark:')
                    #print(X.columns[i])
                    #print('at:')
                    #print(x_cord,y_cord) 
                        
                    
            #2 Här borde sub-bilden sparas med label i en array
            if grid_img_label != 'no_landmark':
                actuallandmarkcounter+=1
                print('grid_img_label: ' + str(grid_img_label))
            grid_image_array.append([grid_img, grid_img_label])
            x_pos += px_x_mov
            #plt.imshow(grid_img, cmap='gray')
            #print(grid_img)
            #return
        x_pos = 0
        y_pos += px_y_mov

    print('Landmark counter: ' + str(land_mark_counter))
    print('Actual Landmark appended: ' + str(actuallandmarkcounter))
#Detta är arrayen med alla sub-bilder och labels
#grid_array = create_grid_boxes(img)
```

```python
#Detta kodblock behöver inte köras, kör kodblocket test() här under
def createModel(training_set, label):
    model = Sequential()
    model.add(tf.keras.layers.Input(shape=(24,32)))
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(800, activation='relu'))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    
    model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=1e-6),
                  loss='binary_crossentropy',
                  metrics=['accuracy'],
                 )
    model.fit(x=training_set, y=label, 
          batch_size = 40, 
          epochs = 50)
    return model

#createModel(training_set, label)
```

```python
def predict_grid_array(img, data):
    px_x_mov = 32
    px_y_mov = 24
    x_max = 640
    y_max = 480
    y_pos = 0
    x_pos = 0
    land_mark_counter = 0
    all_pred_grid_img_arrs = []
    while y_pos < y_max:
        while x_pos < x_max:
            grid_img = img[y_pos:(y_pos+px_y_mov), x_pos:(x_pos+px_x_mov)]
            grid_img_model_pred_val_arr = []
            for model,landmark in models:
                y_pred = model.predict(grid_img, verbose=1)
                #if we predict a landmark we append the prediction score
                if(model.predict_classes(grid_img, verbose=1) == 1):
                    grid_img_model_pred_val_arr.append.([landmark,y_pred[1]]) 
                #else we just append 0.
                else:
                    grid_img_model_pred_val_arr.append([landmark,0]) 
                     
            all_pred_grid_img_arrs.append(grid_img_model_pred_val_arr)
            x_pos += px_x_mov
        x_pos = 0
        y_pos += px_y_mov
        
    return(all_pred_grid_img_arrs) 
```
